```
con = requests.get('https://www.surveymonkey.com/r/?sm=MIYDXxiP3kxT2ECdshqqO')
import re
regex = re.compile('edby="question-title-(\d+)')
fr_question_nums = regex.findall(con.text)
len(fr_question_nums)
```

```
con = requests.get('https://www.surveymonkey.com/r/?sm=MIYDXxiP3kxT')
import re
regex = re.compile('for="(\d+)_(\d+)')
mult_question_nums = regex.findall(con.text)
len(mult_question_nums)
```

```
false_survey_questions = []

for llist in mult_question_nums:
  question = 'survey["' + llist[0] + '[]"].value = ""'
  false_survey_questions.append(question)

for num in fr_question_nums:
  question = 'survey["' + num + '"].value = "'
  false_survey_questions.append(question)

true_survey_questions = []

for llist in mult_question_nums:
  question = 'survey["' + llist[0] + '[]"].value = "' + llist[1] + '"'
  true_survey_questions.append(question)

for num in fr_question_nums:
  question = 'survey["' + num + '"].value = "'
  true_survey_questions.append(question)
```

```
url = 'https://raw.githubusercontent.com/moebot3000/Other_wi/master/forsm1_new.csv?token=AL3PAHXS3LL22BJ6KXZ3S4K5WTHFY'
answers = pd.read_csv(url)
answers.drop(['Unnamed: 0', 'index', 'survey'], axis=1, inplace=True)
```

```
answers
```

```
qsonside = answers.transpose()
qsonside['iffalse'] = false_survey_questions
qsonside['iftrue'] = true_survey_questions
qsonside
```

```
valqsonside = qsonside

testtrues = valqsonside.iloc[0:64,0:65].where(valqsonside == False, valqsonside['iftrue'], axis=0)
testtrues['iffalse'] = valqsonside['iffalse'].iloc[0:64]
mult_testtruefalse = testtrues.iloc[:,0:65].where(testtrues != False, testtrues['iffalse'], axis=0)

for column in mult_testtruefalse:
  test = mult_testtruefalse[column].str.contains('""')
  mult_testtruefalse[column].where(test == False, "", axis=0, inplace=True)
```
